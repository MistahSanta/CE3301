```python
# imports 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("inputOutputData\inputOutput1Bit1Hz.csv", encoding='latin1', skiprows=23)

df.head(5)
```

```python
# Graph for the DAC at 1 Hz 

plt.plot(df['Time (s)'], df['Channel 1 (V)'], 'r-', label='Output Signal', linewidth= 3)
plt.plot( df['Time (s)'], df['Channel 2 (V)'], 'b-', label='Input Signal', linewidth= 3)

# Set the x and y range 

plt.ylim(-0.5, 6)

plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.title('Input and Output Signal of Analog to Digital Converter')
plt.show()
```
