```python
# imports 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv("InputOutputData2Bit\InputOutput2Bit1HzPart1.csv", encoding='latin1', skiprows=22)
data2 = pd.read_csv("InputOutputData2Bit\InputOutput2Bit1HzPart2.csv", encoding='latin1', skiprows=22)

data1.head(5)
```

```python
# Graph for the DAC at 1 Hz 
# This data is messed up. The pink and green line overlap which they should not be. Get better data tbh

plt.plot(data1['Time (s)'], data1['Channel 2 (V)'], 'r-', label='Input Signal', linewidth= 3)
plt.plot(data1['Time (s)'], data1['Channel 1 (V)'], 'b-', label='A0', linewidth= 2)
plt.plot(data2['Time (s)'], data2['Channel 1 (V)'], 'g-', label='A1', linewidth= 2)
plt.plot(data2['Time (s)'], data2['Channel 1 (V)'], 'm-', label='A2', linewidth= 2)

# Set the x and y range 

plt.ylim(-0.5, 6)
plt.xlim(-0.5, 0.6)
plt.legend(loc='upper right', ncol=len(data1.columns) + 1)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.title('Input and Output Signal of Analog to Digital Converter')
plt.show()
```

```python
# Graph for the DAC at 1 Hz 
data3 = pd.read_csv("InputOutputData2Bit\InputOutput2Bit100HzPart1.csv", encoding='latin1', skiprows=22)
data4 = pd.read_csv("InputOutputData2Bit\InputOutput2Bit100HzPart2.csv", encoding='latin1', skiprows=22)


plt.plot(data3['Time (s)'], data3['Channel 2 (V)'], 'r-', label='Input Signal', linewidth= 3)
plt.plot(data3['Time (s)'], data3['Channel 1 (V)'], 'b-', label='A0', linewidth= 2)
plt.plot(data4['Time (s)'], data4['Channel 1 (V)'], 'g-', label='A1', linewidth= 2)
plt.plot(data4['Time (s)'], data4['Channel 2 (V)'], 'm-', label='A2', linewidth= 2)

# Set the x and y range 

plt.ylim(-0.5, 6)
plt.xlim(0.05, 0.07)
plt.legend(loc='upper right', ncol=len(data1.columns) + 1)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.title('Input and Output Signal of Analog to Digital Converter')
plt.show()
```
